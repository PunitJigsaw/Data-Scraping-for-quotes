### Combining everything into function

for extracting Quotes, authorName, Like from all pages

```python
def quotes(soup):
    author_name = [i.text.strip() for  i  in soup.find_all('span', class_='authorOrTitle')]
    quotes = [list(i)[0].strip() for i in soup.find_all('div', class_='quoteText')]
    likes = [i.text.split()[0] for i in soup.find_all('a', class_='smallText')]
    df = pd.DataFrame({'author':author_name, 'quote':quotes, 'likes':likes})
    return df
```

```python
df = quotes(soup)
```

###  Extracting quotes from 2nd, 3rd pages

Can be done by passing each page soup object to quotes() function

 Page:2
 - https://www.goodreads.com/quotes?page=2
 
 
 Page:3
 - https://www.goodreads.com/quotes?page=3

```python
stud = 'https://www.goodreads.com/quotes?page={}'
```

```python
stud.format(2)
```

```python
table = []
for i in range(1,21):
    url_all = stud.format(i)
    html_all = requests.get(url_all).text
    soup = BeautifulSoup(html_all, "html.parser")
    table.append(quotes(soup))
```

```python
# Context: we dont know how much time it take?
# we dont have to wait for time, just calculating makes it easier
tables = []
from tqdm import tqdm
for i in tqdm(range(1,21)):
    url_all = stud.format(i)
    #print(url_all)
    html_all = requests.get(url_all).text
    soup = BeautifulSoup(html_all,'html.parser')
    tables.append(quotes(soup))
```

```python
len(tables)
```

```python
data = pd.concat(tables,axis=0)
data.head()
```

```python
url = 'https://www.goodreads.com/quotes/tag/funny'
html_funny = requests.get(url).text
soup1 = BeautifulSoup(html_funny, "html.parser")
```

```python
soup.find_all('div', class_='greyText smallText left')[0]
```

```python

tags = [i.text.split()[1:] for i in [i for i in soup.find_all('div', class_='greyText smallText left')]]
tags
```

```python
def quotes_funny(soup):
    author_name = [i.text.strip() for  i  in soup.find_all('span', class_='authorOrTitle')]
    quotes = [list(i)[0].strip() for i in soup.find_all('div', class_='quoteText')]
    likes = [i.text.split()[0] for i in soup.find_all('a', class_='smallText')]
    tags = [i.text.split()[1:] for i in [i for i in soup.find_all('div', class_='greyText smallText left')]]
    df = pd.DataFrame({'author':author_name, 'quote':quotes, 'likes':likes, 'tags':tags})
    return df
```
